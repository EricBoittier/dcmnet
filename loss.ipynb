```python
def plot_loss(fn, y_min=0, y_max=0.3):
    with open(fn) as f:
        lines = f.readlines()
    filename = fn
    lines = [_ for _ in lines if _.__contains__("loss")]
    val_loss = [float(_.split()[-1]) for _ in lines]
    train_loss = [float(_.split()[-2]) for _ in lines]
    fn = fn.split("-")[0]
    plt.plot(val_loss, "-", alpha=0.4, label=f"val. {fn} ({min(val_loss)})")
    plt.plot(train_loss, label=f"train {fn} ({min(train_loss)})")
    plt.ylim(y_min, y_max)
    plt.axhline(min(val_loss), alpha=0.4)
    plt.ylabel("Loss [a.u.]", fontsize=20)
    plt.xlabel("Epoch", fontsize=20)
    plt.legend(fontsize=14)
    plt.savefig(f"figures/loss/{filename}.pdf", bbox_inches="tight")
```
